```python
#merge,concatinate and join datasets in pandas 
#concat (outer join, joins everthing and puts them into one table)
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2= pd.DataFrame({"A":[1,22,3,4],"C":[52,62,72,82]})
df3 = pd.concat([df1,df2])
df3
```

```python
#now we have merged all the data in one sinlg etable but if we want to merge the data then we can just pass the values in such a way that they all get merged with common things
df4 = pd.concat([df1,df2],axis=1,join='inner')
#if axis=0 then only the common coulmn will be shown..if axis=1 then all the table wil be shown 
#in this inner join it only added the common coulmn form both the tables..and the common table from both of them was "a"
df4
```

```python
#now we have merged all the data in one sinlg etable but if we want to merge the data then we can just pass the values in such a way that they all get merged with common things
df4 = pd.concat([df1,df2],axis=1,join='outer')
#if axis=0 then only the common coulmn will be shown..if axis=1 then all the table wil be shown 
#in this inner join it only added the common coulmn form both the tables..and the common table from both of them was "a"
df4
```

```python
#merging two datasets
df5 = pd.merge(df1,df2,on=["A"],how='left')
df5
```

```python
#merging two datasets
df5 = pd.merge(df1,df2,on=["A"],how='right')
df5
```

```python
#merging two datasets
df5 = pd.merge(df1,df2,on=["A"],how='outer')
df5
```

```python
#merging two datasets
df5 = pd.merge(df1,df2,on=["A"],how='inner')
df5
```
